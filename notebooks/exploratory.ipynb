# Tweet emotion dataset — quick EDA

Class balance, sample tweets, and a sanity-check call to `predict_emotions` on a couple of lines once the model has been trained.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent))

import matplotlib.pyplot as plt
import pandas as pd

from src import config
from src.data import load_tweets
```

## Class distribution

13 sentiment labels, heavily imbalanced — `neutral`/`worry`/`happiness` dominate.

```python
X, y = load_tweets()
counts = y.value_counts()
print(counts)
counts.plot.bar(figsize=(10, 4), title="Tweets per sentiment")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

## Sample tweets per class

Spot-check that the labels look reasonable.

```python
df = pd.DataFrame({"sentiment": y, "content": X})
for sentiment in counts.head(6).index:
    sample = df[df.sentiment == sentiment]["content"].iloc[0]
    print(f"[{sentiment}] {sample}")
```

## Try the trained model (after running `python -m src.pipeline --train`)

Run the cell below once a model has been persisted to `models/`.

```python
from src.model import load_all, predict_emotions

embedder, label_encoder, classifier = load_all()
samples = [
    "I would like to smash this computer right now",
    "Summer is almost here, I can't wait!",
    "I miss my old friends from school.",
]
for text, emotion in zip(samples, predict_emotions(samples, embedder, classifier, label_encoder)):
    print(f"[{emotion}] {text}")
```
